# Lab 1: Click to Place Markers

This lab introduces the core `ipyleaflet` interaction pattern:

- display a live map in a notebook
- click on the map
- capture the click coordinates
- place a marker at that location
- report the clicked location in a widget

## Learning Goals

- create an `ipyleaflet.Map`
- respond to map click events with `on_interaction`
- add `Marker` objects dynamically
- use `ipywidgets.Output` and `ipywidgets.HTML` for feedback

```python
from ipyleaflet import Map, Marker, LayersControl, WidgetControl, basemaps, basemap_to_tiles
import ipywidgets as widgets
from IPython.display import display
```

## Starter Map

This cell builds the map and a small control panel.
Run it before moving to the click handler.

```python
status = widgets.HTML(value="<b>Status:</b> Click the map to place a marker.")
log = widgets.Output()
clear_button = widgets.Button(description="Clear Markers", button_style="warning")

panel = widgets.VBox([status, clear_button, log])

m = Map(center=(33.9137, -98.4934), zoom=6)
m.add(basemap_to_tiles(basemaps.OpenStreetMap.Mapnik))
m.add(LayersControl())
m.add(WidgetControl(widget=panel, position="topright"))

markers = []

display(m)
```

## Click Handler

This is the key pattern in the whole lab.
When the user clicks on the map:

- extract the latitude and longitude
- create a marker
- add it to the map
- store it in a list so we can remove it later

```python
def handle_map_click(**kwargs):
    if kwargs.get("type") != "click":
        return

    lat, lon = kwargs["coordinates"]
    marker = Marker(location=(lat, lon), draggable=False)
    m.add(marker)
    markers.append(marker)

    status.value = f"<b>Status:</b> Last click at ({lat:.4f}, {lon:.4f})"

    with log:
        print(f"Placed marker #{len(markers)} at ({lat:.4f}, {lon:.4f})")


def clear_markers(button):
    while markers:
        marker = markers.pop()
        m.remove(marker)

    status.value = "<b>Status:</b> All markers cleared."
    with log:
        print("Cleared all markers.")


m.on_interaction(handle_map_click)
clear_button.on_click(clear_markers)
```
