```text
project/
```

If the notebook runs from `project/`, then the file path to `countries.geojson` is:

```python
Path("data/countries.geojson")
```

````python
from pathlib import Path

data_file = Path("data/countries.geojson")

print("Path being checked:", data_file)
print("Exists:", data_file.exists())
````

## Why Check Existence First?

Checking whether a file exists helps students debug path mistakes before trying to load the file.

That means fewer crashes and fewer emotional speeches directed at the laptop.

````python
possible_files = [
    Path("data/countries.geojson"),
    Path("data/missing.geojson"),
    Path("output"),
]

for p in possible_files:
    print(f"{p!s:<28} exists={p.exists()}  is_file={p.is_file()}  is_dir={p.is_dir()}")
````

## Optional Advanced — Set Up a Local Workspace

Use `pathlib` (no shell commands) to:

1. Create a `data/` folder inside `00-Paths/` using `Path.mkdir(exist_ok=True)`
2. Create an `output/` folder the same way
3. Copy one json file from the module's data folder into your new local `data/` using `shutil.copy2`
4. Re-run the existence check from above and confirm all three paths are found

Hint: `Path.mkdir(parents=True, exist_ok=True)` will not error if the folder already exists.

````python
from pathlib import Path
import shutil

local_data = Path("data")
output_dir = Path("output")
local_data.mkdir(exist_ok=True)
output_dir.mkdir(exist_ok=True)

source = Path("../01-JSON_GeoJSON/data/meteorites.geojson")
destination = local_data / source.name
shutil.copy2(source, destination)

print("Copied:", source.resolve())
print("To:    ", destination.resolve())
for path in [local_data, output_dir, destination]:
    print(f"{path} -> exists={path.exists()} is_file={path.is_file()} is_dir={path.is_dir()}")
````
