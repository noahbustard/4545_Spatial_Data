## 02 - Viewing Geojson

### Goal:

Load a geojson object and display it on a map layer.

### Background:

Geojson as we know is a collection of features. The features are accessed using the `features` key which points to a list of json objects.

**Remember:**

- `[]` = list or array of items
- `{}` = object or key value pairs

### Feature Properties

Each feature in the list contains a key: `properties` that points to some key value pairs that dictate some information and styling for that particular object (feature). The following example gives a name to the feature, and defines marker to be placed at a point type. 

```json
"properties": {
        "Name": "Bolin Hall",
        "marker-color": "#00ff00",
        "marker-size": "large",
        "marker-symbol": "zoo",
      }
```

Not all properties apply to every feature type. For example, a marker won't show up if in the properties of a polygon, but the properties below would effect a polygon. In this case it would fill the polygon with pink, outline it with blue, and make the fill extremely see through (alpha channel). Opacity is a value from 0-1 where 0 = clear and 1 = solid. 

```json
"fill": "#ff00ff",
"fill-opacity": 0.2,
"stroke": "#0000ff"
```

### Run the Code

If you run the code below, you should see the geojson file get displayed on a map layer. 

````python
import json
import folium
from rich import print

msu = {"name": "MSU Texas", "lat": 33.87372, "lon": -98.51947}
restaurant = {"name": "Whataburger", "lat": 33.8786, "lon": -98.52275}

route_coordinates = [
    [msu["lon"], msu["lat"]],
    [-98.5206, 33.8751],
    [-98.5217, 33.8766],
    [restaurant["lon"], restaurant["lat"]],
]

gj = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "properties": {
                "Name": "Bolin Hall",
                "marker-color": "#2a9d8f",
                "marker-size": "large",
                "marker-symbol": "college",
            },
            "geometry": {
                "type": "Point",
                "coordinates": [msu["lon"], msu["lat"]],
            },
            "id": 0,
        },
        {
            "type": "Feature",
            "properties": {
                "name": "Campus study area",
                "fill": "#f4a261",
                "fill-opacity": 0.28,
                "stroke": "#264653",
                "stroke-width": 3,
            },
            "geometry": {
                "type": "Polygon",
                "coordinates": [
                    [
                        [-98.5233, 33.8796],
                        [-98.5233, 33.8710],
                        [-98.5165, 33.8710],
                        [-98.5165, 33.8796],
                        [-98.5233, 33.8796],
                    ]
                ],
            },
            "id": 1,
        },
        {
            "type": "Feature",
            "properties": {
                "name": restaurant["name"],
                "marker-color": "#e63946",
                "marker-size": "large",
                "marker-symbol": "restaurant",
            },
            "geometry": {
                "type": "Point",
                "coordinates": [restaurant["lon"], restaurant["lat"]],
            },
            "id": 2,
        },
        {
            "type": "Feature",
            "properties": {
                "name": "Campus lunch route",
                "stroke": "#1d3557",
                "stroke-width": 5,
                "stroke-opacity": 0.85,
            },
            "geometry": {
                "type": "LineString",
                "coordinates": route_coordinates,
            },
            "id": 3,
        },
    ],
}

print("GeoJSON type:", gj["type"])
print("Feature count:", len(gj["features"]))
print(gj)

m = folium.Map(location=[msu["lat"], msu["lon"]], zoom_start=16, tiles="CartoDB positron")

folium.GeoJson(gj, name="msu_features").add_to(m)
folium.Marker(
    location=[msu["lat"], msu["lon"]],
    tooltip=msu["name"],
    icon=folium.Icon(color="green", icon="graduation-cap", prefix="fa"),
).add_to(m)
folium.Marker(
    location=[restaurant["lat"], restaurant["lon"]],
    tooltip=restaurant["name"],
    icon=folium.Icon(color="red", icon="cutlery", prefix="fa"),
).add_to(m)
folium.LayerControl().add_to(m)

m
````
