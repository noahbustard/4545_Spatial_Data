# 00 — My Working Directory

Before we do anything spatial, we must know:

- What is the current working directory?
- Does the `data/` folder exist?
- If the data doesn't exist, then stop nicely (don't crash). 
- We can fix it this time by creating and copying whats needed to your local folder.

```python
from pathlib import Path

# Get path of the current working directory
cwd = Path.cwd()
print("Current Working Directory:")
print(cwd)

first_try = cwd / "data"
print("\nFirst place I looked for data:")
print(first_try)
print("Exists there?", first_try.exists())

# The notebook lives in _micro_lessons, so the assignment data folder is one level up.
data_path = first_try if first_try.exists() else cwd.parent / "data"
print("\nUsing data folder at:")
print(data_path)

exists = data_path.exists()
print("\nDoes data folder exist?")
print(exists)

assert exists, (
    f"\n❌ ERROR: 'data/' folder not found at:\n{data_path}\n"
    "Check your working directory or update the path to find your data folder."
)

print("\n✅ Environment looks correct.")
```
