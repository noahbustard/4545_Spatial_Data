# 🎯 Notebook 04: Indexing and Selection

Time to slice and dice.

In this notebook, you'll:
- Use `.loc[]` for label-based selection
- Use `.iloc[]` for position-based selection
- Create boolean masks to filter rows
- Select specific columns and rows (solo or together)

Let the chaos of index manipulation begin.
---

```python
import pandas as pd
from pathlib import Path

# Use the repo-local Titanic CSV so the notebook runs without seaborn/network access.
DATA_PATH = Path('Assignments/01-Python_Foundations/05-Working_with_Data_Adv/data/titanic.csv')
if not DATA_PATH.exists():
    DATA_PATH = Path('../../data/titanic.csv')

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip().str.lower()
df['class'] = df['pclass'].map({1: 'First', 2: 'Second', 3: 'Third'})
df.head(3)
```

## 🔎 Select rows by label with `.loc[]`

```python
# .loc[row_label, column_label]
# Titanic uses default RangeIndex, so labels are integers.
print(df.loc[0])

# Row ranges in .loc are inclusive when the index labels are integers.
df.loc[0:4, ['age', 'fare', 'sex']]
```

## 🔢 Select rows by position with `.iloc[]`

```python
# .iloc uses pure positions.
print(df.iloc[0])

# Row slice, column slice. The stop value is exclusive.
df.iloc[1:4, 0:5]
```

## 🎯 Boolean Masking (Filtering)

```python
# Filter passengers under age 10.
kids = df[df['age'] < 10]
kids[['name', 'age', 'sex', 'class']].head()
```

## 🧠 Compound Conditions

```python
# Adults in First Class who paid over $100.
first_class_high_fare = df[(df['age'] > 18) & (df['class'] == 'First') & (df['fare'] > 100)]
first_class_high_fare[['name', 'age', 'class', 'fare']].head()
```

## 🧹 Select Columns Only

```python
df[['sex', 'age', 'embarked']].head()
```

## 🎯 Your Turn

1. Select the first row using `.iloc[]`, then again using `.loc[]`.
2. Slice the first 5 rows and select only `survived`, `sex`, and `age` columns.
3. Filter the dataset for female passengers over 40.
4. Create a DataFrame of passengers in 2nd class under age 15.
5. Use `.loc[]` to select the rows 10–20 and only the columns `fare`, `class`, and `age`.

🎯 **Bonus:** Find passengers with missing `age` and print only their `name`, `age`, and `pclass` (use `.isnull()`).

```python
# 1. Select the first row using .iloc[], then again using .loc[].
first_row_by_position = df.iloc[0]
first_row_by_label = df.loc[0]
print('Same first passenger:', first_row_by_position['name'] == first_row_by_label['name'])

# 2. Slice the first 5 rows and select only survived, sex, and age columns.
first_five = df.loc[0:4, ['survived', 'sex', 'age']]
print()
print('First five selected columns:')
print(first_five)

# 3. Filter the dataset for female passengers over 40.
female_over_40 = df[(df['sex'] == 'female') & (df['age'] > 40)]
print()
print('Female passengers over 40:', len(female_over_40))
print(female_over_40[['name', 'age', 'fare']].head())

# 4. Create a DataFrame of passengers in 2nd class under age 15.
second_class_children = df[(df['pclass'] == 2) & (df['age'] < 15)]
print()
print('Second-class passengers under 15:', len(second_class_children))
print(second_class_children[['name', 'age', 'class']].head())

# 5. Use .loc[] to select rows 10-20 and only fare, class, and age.
rows_10_to_20 = df.loc[10:20, ['fare', 'class', 'age']]
print()
print('Rows 10-20, selected columns:')
print(rows_10_to_20)

# Bonus: passengers with missing age.
missing_age = df.loc[df['age'].isnull(), ['name', 'age', 'pclass']]
print()
print('Passengers with missing age:', len(missing_age))
missing_age.head()
```
