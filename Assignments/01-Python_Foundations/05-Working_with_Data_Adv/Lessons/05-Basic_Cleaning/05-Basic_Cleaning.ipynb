# 🧼 Cleaning Dirty Data (Missing Values & Type Fixes)

## 🔹 LEARNING GOALS:
- Detect and count missing values (`NaN`)
- Fill or drop missing data
- Convert column data types safely
- Understand the difference between `NaN`, `None`, `""`, and type mismatches


### 🧪 1. Load a Messy Dataset

```python
import pandas as pd
import numpy as np
from pathlib import Path

# A deliberately messy table: missing values, blank strings, invalid numbers,
# invalid dates, and one extra column with bad email values.
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None, '???'],
    'Age': ['25', 'thirty', 35, np.nan, '40', 'old'],
    'Signup Date': ['2022-01-01', 'not a date', '2022/03/01', None, 'April 5, 2022', 'not sure'],
    'Score': [95.5, None, 88.0, 92.5, '', 'A+'],
    'Email': ['alice@example.com', None, '', 'david@example.com', 'missing@example.com', 'bad-email'],
}

df = pd.DataFrame(data)
df
```

### 🧯 2. Detecting Missing or Broken Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

### 🧹 3. Cleaning Strategy Options

```python
df.fillna({
    'Name': 'Unknown',
    'Age': -1,
    'Signup Date': '1970-01-01',
    'Score': 0.0,
    'Email': 'unknown@example.com',
})
```

```python
df.dropna()
```

### 🧬 4. Data Type Fixes

```python
df.dtypes
```

```python
# Convert blank strings to NaN first, then coerce invalid numbers/dates to NaN.
df = df.replace('', np.nan)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
df['Signup Date'] = pd.to_datetime(df['Signup Date'], errors='coerce')
df.dtypes
```

### 🩹 5. Impute (Fill In) Fixed Missing Values

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Score'] = df['Score'].fillna(df['Score'].mean())
```

```python
df['Signup Date'] = df['Signup Date'].fillna(df['Signup Date'].min())
```

```python
df['Name'] = df['Name'].fillna('Unknown')
df['Email'] = df['Email'].fillna('unknown@example.com')
df
```

### 🤓 6. Cleaned Data Review

```python
df.info()
```

```python
df.describe(include="all")
```

### 🧠 Mini-Challenge

> 🗂 Load `"data/survey.csv"` and:
- Identify which columns have missing values
- Use `.isnull().sum()` to get a null report
- Use a mix of `.fillna()`, `.dropna()`, and `pd.to_numeric()` or `pd.to_datetime()` to clean it
- Print a summary with `.info()` and `.describe()`

```python
# Mini-challenge: load and clean the messy survey data.
SURVEY_PATH = Path('Assignments/01-Python_Foundations/05-Working_with_Data_Adv/data/survey.csv')
if not SURVEY_PATH.exists():
    SURVEY_PATH = Path('../../data/survey.csv')

survey = pd.read_csv(SURVEY_PATH)
print('Original shape:', survey.shape)
print()
print('Missing values before cleaning:')
print(survey.isnull().sum().sort_values(ascending=False).head(10))

clean = survey.copy()

# Standardize obvious text categories.
clean['major'] = clean['major'].replace({'CompSci': 'Computer Science', 'CS': 'Computer Science'})
clean['gender'] = clean['gender'].fillna('Unknown').str.strip().str.title()
clean['dorm_resident'] = clean['dorm_resident'].fillna('Unknown').str.strip().str.title()
clean['favorite_language'] = clean['favorite_language'].fillna('Unknown').str.strip()
clean['comments'] = clean['comments'].fillna('No comment')

# Convert numeric-looking columns and mark impossible values as missing.
numeric_columns = ['age', 'gpa', 'study_hours_per_week', 'satisfaction_1_5', 'expected_grad_year', 'commute_minutes']
for column in numeric_columns:
    clean[column] = pd.to_numeric(clean[column], errors='coerce')

clean.loc[~clean['age'].between(16, 80), 'age'] = np.nan
clean.loc[~clean['gpa'].between(0, 4), 'gpa'] = np.nan
clean.loc[clean['study_hours_per_week'] < 0, 'study_hours_per_week'] = np.nan
clean.loc[~clean['satisfaction_1_5'].between(1, 5), 'satisfaction_1_5'] = np.nan
clean.loc[clean['commute_minutes'] < 0, 'commute_minutes'] = np.nan

# Attendance rate has percent strings like '105%'. Convert to a 0-100 numeric rate.
clean['attendance_rate'] = clean['attendance_rate'].astype(str).str.rstrip('%')
clean['attendance_rate'] = pd.to_numeric(clean['attendance_rate'], errors='coerce')
clean.loc[~clean['attendance_rate'].between(0, 100), 'attendance_rate'] = np.nan

# Convert dates. Invalid dates become NaT.
clean['survey_date'] = pd.to_datetime(clean['survey_date'], errors='coerce')

# Fill remaining values with simple course-appropriate defaults.
for column in ['age', 'gpa', 'study_hours_per_week', 'satisfaction_1_5', 'commute_minutes', 'attendance_rate']:
    clean[column] = clean[column].fillna(clean[column].median())

clean['expected_grad_year'] = clean['expected_grad_year'].fillna(clean['expected_grad_year'].mode()[0])
clean['income_range'] = clean['income_range'].fillna('Unknown')
clean['clubs_member'] = clean['clubs_member'].fillna('Unknown')
clean['survey_date'] = clean['survey_date'].fillna(clean['survey_date'].min())

print()
print('Missing values after cleaning:')
print(clean.isnull().sum().sort_values(ascending=False).head(10))
print()
print('Cleaned info:')
clean.info()

clean.describe(include='all')
```
