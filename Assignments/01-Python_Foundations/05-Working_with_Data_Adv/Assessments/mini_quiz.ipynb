# Working with Data Advanced Mini Quiz

Completed assessment using pandas with the repo-local Titanic CSV.

```python
from pathlib import Path
import pandas as pd

DATA_PATH = Path('Assignments/01-Python_Foundations/05-Working_with_Data_Adv/data/titanic.csv')
if not DATA_PATH.exists():
    DATA_PATH = Path('../data/titanic.csv')

df = pd.read_csv(DATA_PATH)
df.head()
```

## Q1. Shape and columns

```python
print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
assert df.shape[0] == 891
```

## Q2. Select and filter rows

```python
first_class = df[df['Pclass'] == 1]
high_fare = df.loc[df['Fare'] > 100, ['Name', 'Pclass', 'Fare']]
print('First-class passengers:', len(first_class))
print('High-fare passengers:', len(high_fare))
high_fare.head()
```

## Q3. Missing values and summary statistics

```python
missing = df.isna().sum().sort_values(ascending=False)
print(missing.head())
summary = df[['Age', 'Fare', 'Survived']].describe()
summary
```

## Q4. Grouping

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean().rename('survival_rate')
survival_by_class
```

## Reflection

The Titanic data shows why inspection matters: `Age` and `Cabin` have missing values, fares are skewed, and survival rate changes by passenger class.
