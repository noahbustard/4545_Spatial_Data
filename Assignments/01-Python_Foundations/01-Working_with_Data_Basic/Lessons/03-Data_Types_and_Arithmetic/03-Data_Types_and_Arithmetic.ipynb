# 📘 03 - Data Types and Arithmetic

Explore Python basics like variables, data types, and arithmetic operations.

## ✅ Learning Goals
- Understand Python data types
- Use variables and print statements
- Perform basic arithmetic

```python
# 🧠 Variables and Types
x = 10
y = 3.5
z = 'hello'
print(type(x), type(y), type(z))
```

```python
# ➕ Arithmetic Operations
print('Add:', x + y)
print('Divide:', x / y)
print('Floor Divide:', x // y)
print('Exponent:', x ** 2)
```

## ✏️ Try It Yourself
- Create your own variables
- Try combining numbers and strings
- Use `input()` to interact

```python
favorite_number = 17
pi_guess = input('What is your guess for the value of pi? ')
city = 'Wichita Falls'

print(type(favorite_number), favorite_number)
print(type(pi_guess), pi_guess)
print(type(city), city)
print('Favorite number squared:', favorite_number ** 2)
print('City banner:', city + ' Rocks')
```
