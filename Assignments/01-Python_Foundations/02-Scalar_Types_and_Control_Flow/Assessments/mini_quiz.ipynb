## 5️⃣ Given the following:

```Python
a = 11

if a % 2 == 0:
    print("Variable is even")
```

Add to the if statement so that it will also print "Variable is odd" depending on the number.

````python
a = 11

if a % 2 == 0:
    print('Variable is even')
else:
    print('Variable is odd')
````
