## Q2. File I/O check

Write text to a temporary file, read it back, and confirm the contents match.

```python
from pathlib import Path
import tempfile

with tempfile.TemporaryDirectory() as tmpdir:
    path = Path(tmpdir) / 'hello.txt'
    path.write_text('Hello World', encoding='utf-8')
    contents = path.read_text(encoding='utf-8')

print(contents)
assert contents == 'Hello World'
```

## Q3. Lists, dictionaries, and loops

Build a small grade book and calculate an average score.

```python
scores = {'Ada': 96, 'Grace': 91, 'Katherine': 98}
average = sum(scores.values()) / len(scores)
for name, score in scores.items():
    print(f'{name}: {score}')
print(f'Average: {average:.1f}')
```

## Q4. Plotting basics

Create a simple bar chart from the same grade book.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(scores.keys(), scores.values(), color='#4f46e5')
ax.set_title('Mini Quiz Scores')
ax.set_ylabel('Score')
ax.set_ylim(0, 105)
plt.show()
```
