# Describing and Visualizing Data Mini Quiz

Completed assessment using numeric summaries and matplotlib visualizations.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

DATA_PATH = Path('Assignments/01-Python_Foundations/05-Working_with_Data_Adv/data/pokemon.csv')
if not DATA_PATH.exists():
    DATA_PATH = Path('../../05-Working_with_Data_Adv/data/pokemon.csv')

pokemon = pd.read_csv(DATA_PATH)
pokemon.head()
```

## Q1. Identify data types

```python
print(pokemon[['Name', 'Type 1', 'Total', 'HP', 'Attack', 'Legendary']].dtypes)
print('Rows:', len(pokemon))
```

## Q2. Summary statistics

```python
stats = pokemon[['Total', 'HP', 'Attack', 'Defense', 'Speed']].describe()
stats
```

## Q3. Categorical comparison

```python
type_counts = pokemon['Type 1'].value_counts().head(10)
type_counts
```

## Q4. Visualization

```python
fig, ax = plt.subplots(figsize=(7, 4))
type_counts.sort_values().plot(kind='barh', ax=ax, color='#0891b2')
ax.set_title('Top Pokemon Primary Types')
ax.set_xlabel('Count')
ax.set_ylabel('Primary type')
plt.tight_layout()
plt.show()
```

## Reflection

The most common primary types in this dataset are Water and Normal. The summary statistics show a wide spread in total power and individual battle stats.
